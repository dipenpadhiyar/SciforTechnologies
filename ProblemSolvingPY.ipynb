(6)What is Polymorphism in Python? Give an example.
```python
Polymorphism allows objects of different classes to be treated as objects of a common superclass.

class Car:
    def drive(self):
        raise NotImplementedError("Subclasses must implement this method")


class Sedan(Car):
    def drive(self):
        print("Sedan is cruising smoothly")


class SUV(Car):
    def drive(self):
        print("SUV is handling rough terrain")


# Polymorphic behavior
def drive_car(car):
    car.drive()


sedan = Sedan()
suv = SUV()

drive_car(sedan)
drive_car(suv)

```

(15)What is the difference between an instance variable and a class variable?
How do you implement Encapsulation, Abstraction, Polymorphism?
```python
Instance Variable:
-Belongs to an instance of a class.
-Each instance (object) of the class has its own copy.
-Defined inside methods and initialized using self.
-Accessed using instance/object reference (self.variable).

Class Variable:
-Belongs to the class itself.
-Shared among all instances of the class.
-Defined outside methods and within the class body.

from abc import ABC, abstractmethod

class EncapsulationExample:
    class Book:
        def __init__(self, title, author):
            self.__title = title
            self.__author = author

        def get_title(self):
            return self.__title

        def get_author(self):
            return self.__author

    class AbstractionExample(ABC):
        @abstractmethod
        def calculate(self):
            pass

    class Rectangle(AbstractionExample):
        def __init__(self, length, width):
            self.length = length
            self.width = width

        def calculate(self):
            return self.length * self.width

    class PolymorphismExample:
        def sum(self, a, b, c=0):
            return a + b + c

example = EncapsulationExample()

# Encapsulation
book_instance = example.Book(title="Wings of Fire", author="Dr. A. P. J. Abdul Kalam")
print(book_instance.get_title())
print(book_instance.get_author())

# Abstraction
rectangle_instance = example.Rectangle(length=5, width=10)
print(rectangle_instance.calculate())

# Polymorphism
polymorphism_example = example.PolymorphismExample()
print(polymorphism_example.sum(3, 4))
print(polymorphism_example.sum(3, 4, 9))
```

(16)How do you Implement single level Inheritance, multiple level inheritance, multi level inheritance, Hybrid Inheritance
```python
# Single Level Inheritance
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def drive(self):
        print("Car is being driven")

# Multiple Level Inheritance
class Boat:
    def float(self):
        print("Boat is floating")

class Ship(Boat):
    def sail(self):
        print("Ship is sailing")

# Multi Level Inheritance
class Airplane:
    def fly(self):
        print("Airplane is flying")

class Jet(Airplane):
    def turbo(self):
        print("Jet is at turbo speed")

class FighterJet(Jet):
    def attack(self):
        print("Fighter jet is attacking")

# Hybrid Inheritance (Combination of above)
class Hybrid(Car, Ship, FighterJet):
    pass

```
